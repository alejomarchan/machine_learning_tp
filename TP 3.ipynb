# Ejercicio 2

La √∫ltima novela de cierto autor ha tenido un importante √©xito, hasta el punto de que el 80 % de los lectores ya la han le√≠do. Un grupo de cuatro amigos son aficionados a la lectura:

a. ¬øCu√°l es la probabilidad de que en el grupo hayan le√≠do la obra dos personas? ¬øY al menos dos?

b. Calcule la media y la varianza. Interprete los resultados.

## Soluci√≥n
## Definici√≥n de la soluci√≥n
#### Este pertenece a una Distribuci√≥n Binomial
P(X = k) = C(n, k) * p^k * (1-p)^(n-k)

Donde P(X = k) representa la probabilidad de que k personas hayan le√≠do la obra. La variable aleatoria ser√≠a la cantidad de personas que hayan le√≠do la obra.

```python
from scipy.stats import binom

prob_exito = 0.8
num_ensayos = 4

# Probabilidad de exactamente 2 personas hayan le√≠do la obra
prob_2_pers = binom.pmf(2, num_ensayos, prob_exito)

# Probabilidad de al menos dos personas hayan le√≠do la obra
prob_al_menos_2_pers = 1 - binom.cdf(1, num_ensayos, prob_exito)

# Media y varianza
media = num_ensayos * prob_exito
varianza = num_ensayos * prob_exito * (1 - prob_exito)

print("Probabilidad de que en el grupo hayan le√≠do la obra dos personas:", prob_2_pers)
print("Probabilidad de que en el grupo hayan le√≠do la obra al menos dos personas:", prob_al_menos_2_pers)
print("Media:", media)
print("Varianza:", varianza)
```

# Ejercicio 3 
Si 0,05 es la probabilidad de que cierta pieza en una l√≠nea de producci√≥n sufra una desviaci√≥n excesiva. ¬øcu√°l es la probabilidad de que la sexta pieza probados sea la primera en mostrar dicha desviaci√≥n?

## Soluci√≥n
## Definici√≥n de la soluci√≥n
#### Este pertenece a una Distribuci√≥n Geom√©trica
P(X = k) = (1 - p)^(k - 1) * p

P(X = k) es la probabilidad de que el evento ocurra en el ensayo n√∫mero k

```python
from scipy.stats import binom

prob_desv = 0.05
num_piezas = 6
pos_fallo = 1

prob_primera= binom.pmf(pos_fallo, num_piezas, prob_desv)

print("La probabilidad de que la sexta pieza sea la primera en mostrar desviaci√≥n es:", prob_primera)
```

# Ejercicio 4
Una m√°quina de bebidas se regula para que sirva 200 ml por vaso. Si la cantidad de bebida se distribuye normalmente con desvi√≥ est√°ndar de 15 ml.

a. ¬øCu√°l es la probabilidad de que una bebida contenga m√°s de 224 ml?

b. Determine el porcentaje de bebidas que contendr√°n entre 191 ml y 209 ml.

c. ¬øCu√°l es la probabilidad de que una bebida contenga menos de 190 ml?

d. En todos los casos anteriores grafiqu√© la probabilidad como el √°rea debajo de la curva.

## Soluci√≥n
## Definici√≥n de la soluci√≥n
#### Este pertenece a una Distribuci√≥n Normal

Y la variable aleatoria es la cantidad de bebida servida por vaso

```python
#Colocando los datos de Entrada
import scipy.stats as stats

media = 200
desv_std = 15
```

### Parte a

```python
x = 224

prob = 1 - stats.norm.cdf(x, media, desv_std)
print("La probabilidad de que una bebida contenga m√°s de 224 ml es:", prob)
```

### Parte b

```python
x1 = 191
x2 = 209

prob_rango = stats.norm.cdf(x2, media, desv_std) - stats.norm.cdf(x1, media, desv_std)
porc = prob_rango * 100
print("El porcentaje de bebidas que contendr√°n entre 191 ml y 209 ml es:", porc)
```

### Parte c

```python
x3 = 190

prob_menor = stats.norm.cdf(x3, media, desv_std)
print("La probabilidad de que una bebida contenga menos de 190 ml es:", prob_menor)
```

### Parte d

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x_values = np.linspace(media - 3 * desv_std, media + 3 * desv_std, 100)

# Calculate y values using PDF
y_values = stats.norm.pdf(x_values, media, desv_std)

# Plot the PDF
plt.plot(x_values, y_values)

# Shade the area for the desired probabilities
plt.fill_between(x_values, y_values, where=(x_values > x), alpha=0.5, color='red')
plt.fill_between(x_values, y_values, where=((x1 < x_values) & (x_values < x2)), alpha=0.5, color='green')
plt.fill_between(x_values, y_values, where=(x_values < x3), alpha=0.5, color='blue')

# Set labels and title
plt.xlabel('Cantidad de bebida (ml)')
plt.ylabel('Probabilidad')
plt.title('Distribuci√≥n Normal de la Cantidad de Bebida')

# Show the plot
plt.show()
```
